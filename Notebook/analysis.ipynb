```python
import pandas as pd

# 1. Fresh load
df = pd.read_csv("../Data/Nassau Candy Distributor.csv")

# 2. Convert dates (DD-MM-YYYY)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y', errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y', errors='coerce')

# 3. Fix wrong Ship Date year using Order Date year
df['Ship Date'] = df.apply(
    lambda row: row['Ship Date'].replace(year=row['Order Date'].year)
    if pd.notnull(row['Order Date']) and pd.notnull(row['Ship Date'])
    else pd.NaT,
    axis=1
)

# 4. Create Delivery Days
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# 5. Extract useful date columns
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day

# 6. Check result
df[['Order Date', 'Ship Date', 'Delivery Days', 'Order Year', 'Order Month', 'Order Day']].head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# 1) Load data
df = pd.read_csv("../Data/Nassau Candy Distributor.csv")

# 2) Convert dates (DD-MM-YYYY)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

# 3) Create Delivery Days
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# 4) Extra useful columns
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day

# 5) Quick check
df[['Order Date', 'Ship Date', 'Delivery Days', 'Order Year', 'Order Month', 'Order Day']].head()
```
